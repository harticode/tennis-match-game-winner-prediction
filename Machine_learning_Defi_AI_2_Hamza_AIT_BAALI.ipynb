## **I.** 🎾 **Understand Tennis**

### **The Framework**
Each tennis match is made up of two to three sets. To win a set, you must win at least six games.
The games are scored starting at "love" (or zero) and go up to 40, but that's actually just four points. From love, the first point is 15, then 30, then 40, then game point, which wins the game.

### **Starting the Game**
To determine who serves first, you flip a coin or (more likely) spin a racket. Whoever wins the toss gets to decide one of four things: that she wants to serve first, that she wants to receive first, which side of the court she wants to start on (in which case, the opponent chooses who serves first), or that she wants to leave the choices up to her opponent.
Whoever starts serving continues to serve until that game is over. Then the serve moves to the other player.
You serve from behind the baseline, starting on the right-hand side of the court, anywhere between the singles sideline and the centre mark on the court.

### **Scoring the Game**
Before serving, you should announce the score, with your score first, then your opponent's. So if you have zero and your opponent has 30, say "love-30."
Every time you serve, you get two tries. The serve must go over the net, land in the service box opposite you, and bounce once before your opponent returns it.
If it doesn't land in the service box, you take a second serve. If the second serve also misses (a double fault), then you lose the point.
If your serve grazes the net but still lands in the service box, the serve doesn't count, and you must take that serve over. This is called a "let."
If your serve goes in and the opponent returns it, you continue hitting back and forth until someone hits the ball into the net, hits it out of bounds, or misses a shot. If that person is you, then your opponent gets a point. If it's your opponent who hits it into the net/hits it out of bounds/misses a shot, then you get the point.
Whoever is serving continues serving until the score reaches 40, calling out the score before every serve.
If the score is tied at 40 ("40-all"), that is "deuce," which is essentially another word for tie. To break the tie, someone must win two points in a row. If you are serving at deuce and you win the next point, you get and "advantage". If your opponent wins, it goes back to deuce, and someone again must win by two points in a row. Yes, this could go on forever.
Once the game is over, the other person serves. After odd-numbered games (so after game one, game three, game five) you switch sides on the court.

### **Scoring a Set**
Before the first serve in each new game, whoever is serving announces the score in sets. Say your score first, then your opponent's. So if you won the first set, you would say "1-0."
To complete a set, someone must win six games; the first person to win six games wins the set.
However, as with "deuce," you must win a set by at least two games. So, if the score is 6-5, the person with 5 must win by two games. If the score ties at 6-6, you play a tiebreaker.

### **Scoring the Match**
The whole shebang is called a match. The match is determined by the best two out of three sets. So if you win two sets, you win. If you each win a set, then you play a third set to determine the winner.

### **POINT SYSTEM**
Tennis has a different point system than most sports. Before we go into detail, here is your guide to scoring a game:
* 0 points= Love
* 1 point = 15
* 2 points= 30
* 3 points= 40
* Tied score= All
* 40-40 = Deuce
* Server wins deuce point = Ad-In
* Receiver wins deuce point = Ad-Out

### Place of the Serve

##### **The Three Areas**
* **Wide** – to serve out wide is very popular especially in singles tennis, by serving in this direction it opens up the court as it makes your opponent run to the ball. However, it does give your opponent more options for his return. If you are playing doubles tennis by serving out wide it give your partner a good opportunity to approach the net and anticipate the return.
* **Down the T** – if you prefer playing percentage tennis then serving “down the T” or in the middle does not give your opponent as many angles to hit their service return. By serving down the middle it can limit any cross court returns, and in doubles that gives your partner an opportunity to get into the point. The only real risk of this serve it is quite a small area you are aiming for.
* **Into the Body** – a really effective serve in singles and doubles as it ties your opponent up and does not allow him time to do little but get the ball back. In singles and doubles it often causes short returns therefore setting the server side up for a volley on their next shot. For the server it makes the service easier to produce as there is a bigger area to aim for.

```python
#Separate the data into server data and receiver data, as well as splitting the services

df_features = pd.DataFrame()
df_features['Server'] = df['Server']
#Separate Sets points : Sets1 : server , Sets2: Reciever
df_features[['Sets1','Sets2']] = df['Sets'].str.split('?',expand=True)

#Separate Games points : Games1 : server , Games2: Reciever
df_features[['Games1','Games2']] = df['Games'].str.split('?',expand=True)

#Separate Points : Points1 : server , Points2: Reciever
df_features[['Points1','Points2']] = df['Points'].str.split('?',expand=True)
```

```python
#Normalize Points to get a more significatif value
PointsNormalized = {'0':0,'15':1,'30':2,'40':3,'AD':4}
df_features['Points1Normalized']=[PointsNormalized[x] for x in df_features['Points1']]
df_features['Points2Normalized']=[PointsNormalized[x] for x in df_features['Points2']]
```

### **Features Extraction**

At this point, I will try to extract as much features as I find from this data.
Most of the information are in Description column.

```python
#Description : Features Extraction
#Seperate by Service
df_features[['1stService','2ndService','Rally']] = df['Description'].str.split('.',expand=True)
df_features['Rally']=[df_features['2ndService'][i] if 'rally' in df_features['2ndService'][i]  else x.Rally for i,x in df_features.iterrows() ]
df_features['2ndService']=[x if 'rally' not in x else '' for x in df_features['2ndService'] ]
```

```python
df_features[['1stService','2ndService','Rally']].info()
```

* Based on this information we might not use the Rally information provided in the description. And because we have all the backs and returns in detail, we will try to calculate the Rally length and drop the rally column we have now.

```python
df_features=df_features.drop(['Rally'],'columns')
df_features.head(3)
```

⛳ Now we will try to calculate The Rally length but to do that we need to find the relevant Service, because the 1st Service not always successful. This will help us Extract more informations.


```python
#Get the Relevant Service(The the service that made the diffrence) and number of Services in a Point
df_features['RelevantService']=[x['1stService'] if x['2ndService']=='' else x['2ndService'] for i,x in df_features.iterrows()]
df_features['ServicesNumber']=[1 if x['2ndService']=='' else 2 for i,x in df_features.iterrows()]
df_features.head(2)
```

```python
print("*****************************************************************")
print("***                      Rally Length                         ***")
print("*****************************************************************")
#Rally Length : count of the sequence of back and forth shots between players, within a point.
df_features['RallyLength']=[len(x.split(';')) -1 for x in df_features['RelevantService']]
df_features.info()
```

```python
print("*****************************************************************")
print("***                   More Features Ext                       ***")
print("*****************************************************************")

#(Cushion Point)game is tied — meaning, each player has scored three times
df_features['PointsCushion']=[x.Points1Normalized-x.Points2Normalized if (x.Points1!='AD' and x.Points2!='AD') else 1 for i,x in df_features.iterrows()]  

# Point ended by ('forced error', 'unforced error', 'winner', 'double fault', 'ace', 'service winner'):
df_features['FinishType']=[x.split(',')[-1].replace('\xa0','') for x in df_features['RelevantService']]

#Services Type (' to body', ' wide', ' down the T'): 
df_features['1stServiceType']=[x.split(';')[0].split(',')[0].split('serve')[-1].replace('\xa0','') for x in df_features['1stService']]
df_features['2ndServiceType']=[x.split(';')[0].split(',')[0].split('serve')[-1].replace('\xa0','') for x in df_features['2ndService']]
df_features['RelevantServiceType']=[x.split(';')[0].split(',')[0].split('serve')[-1].replace('\xa0','') for x in df_features['RelevantService']]

df_features.head(2)
```

```python
#Faults in Point (NB : for the server and reciever)
df_features['Fault']=[0 if x=='' else 2 if 'fault' in x else 1 for x in df_features['2ndService']]

#Error in Point (NB : for the server and reciever) : Useless in our case
df_features['Error']=[1 if 'fault' in x else 1 if 'error' in x else 0 for x in df_features.RelevantService]

#who won the point
opponent={'RogerFederer':'RafaelNadal','RafaelNadal':'RogerFederer'}
df_features['PointWinner']=[x.Server if (x.RallyLength%2 + x.Error !=1) else opponent[x.Server.replace('\xa0','')] for i,x in df_features.iterrows()]
df_features['PointWinner']=[x.PointWinner.replace('\xa0','') for i,x in df_features.iterrows()] #Clean Up
```

```python
#Server Wins the Point
df_features['ServerWins']=[1 if x.PointWinner[0:2]==x.Server[0:2] else 0 for i,x in df_features.iterrows()]

#Game Point means the last Point in the game that determine who won the game
df_features['GamePoint']=[1 if ((x.Points1=='40' or x.Points2=='40') and x.PointsCushion!=0) else 0 for i,x in df_features.iterrows()]

#The receiving player can win the game by scoring the next point also
df_features['BreakPoint']=[1 if x.GamePoint==1 and x.PointsCushion <0 else 0 for i,x in df_features.iterrows()]

#The receiving player won the game by a Breakpoint
df_features['Break']=[1 if x.BreakPoint == 1 and x.ServerWins == 0 else 0 for i,x in df_features.iterrows()]

#The receiving player had that number of breakpoints at some time in the game, but the person serving still managed to win.
df_features['BreakSaved']=[1 if x.BreakPoint == 1 and x.ServerWins == 1 else 0 for i,x in df_features.iterrows()]
```

```python
#Convert some Dtype to Int64
df_features["Sets1"] = pd.to_numeric(df_features["Sets1"])
df_features["Sets2"] = pd.to_numeric(df_features["Sets2"])
df_features["Games1"] = pd.to_numeric(df_features["Games1"])
df_features["Games2"] = pd.to_numeric(df_features["Games2"])

#clean Up Server
df_features["Server"] = [x.Server.replace("\xa0", "") for i,x in df_features.iterrows()]
df_features.info()
```

### **Markov Model** (Prediction server winning a single game by point)

I choosed by game and not by Match because our dataset is very limited even for Games, and I see that prediction by Game will result in by Set and than by Match.(the First methode we will use is based on statistics and pure probability)

- calculate the probability of server winning a single game, 
- given p(winning single point) and current point score

```python
# Basic Math Functions
def fact(x):
    if x in [0, 1]:  return 1
    r = 1
    for a in range(1, (int(x)+1)):  r = r*a
    return r
 
def ch(a, b):
    return fact(a)/(fact(b)*fact(a-b))
```

```python
def gameProbability(s, a, b):
    return ch((a+b), a)*(s**a)*((1-s)**b)*s
```

```python
def gameProb(s, s_p1=0, s_p2=0):
    ## s_p1, s_p2 = current game score, where love = 0, 15 = 1, etc.
    ## - e.g. 30-15 is s_p1=2, s_p2=1

    ## check if game is already finished:
    if s_p1 >= 4 and (s_p1-s_p2) >= 2:
        return 1
    elif s_p2 >= 4 and (s_p2-s_p1) >= 2:
        return 0
    else:   pass
    ## if deuce or ad score e.g. 5-4, reduce to e.g. 3-2
    while True:
        if (s_p1+s_p2) > 6:
            s_p1 -= 1
            s_p2 -= 1
        else:   break
    ## specific probabilities:
    if s_p2 == 0:  s_p20 = gameProbability(s, 3-s_p1, 0)
    else:   s_p20 = 0
    if s_p2 <= 1:  s_p215 = gameProbability(s, 3-s_p1, 1-s_p2)
    else:   s_p215 = 0
    if s_p2 <= 2:  s_p230 = gameProbability(s, 3-s_p1, 2-s_p2)
    else:   s_p230 = 0
    if s_p1 == 4:
        s_p2Ad, lAd = s, 0
        d = 1-s
    elif s_p2 == 4:
        s_p2Ad, lAd = 0, 1-s
        d = s
    else:
        s_p2Ad, lAd = 0, 0
        a = 3 - s_p1
        b = 3 - s_p2
        d = ch((a+b), a)*(s**a)*((1-s)**b)
    if s_p1 <= 2:  l30 = gameProbability((1-s), 3-s_p2, 2-s_p1)
    else:   l30 = 0
    if s_p1 <= 1:  l15 = gameProbability((1-s), 3-s_p2, 1-s_p1)
    else:   l15 = 0
    if s_p1 == 0:  l0 = gameProbability((1-s), 3-s_p2, 0)
    else:   l0 = 0
    ## given d = prob of getting to deuce,
    ## math to divide up further gameProbability
    dv = s**2 + (1-s)**2
    s_p2d = (d*(s**2))/dv
    ld = (d*((1-s)**2))/dv
    win = s_p20 + s_p215 + s_p230 + s_p2d + s_p2Ad
    lose = l0 + l15 + l30 + ld + lAd
    return win
```

## Fine-tune Random Forest : GameWinner Label Encoding
in this task we are trying to find the best Random Forest Model, basically our Idea is to try diffrent sets of parameters till we find the best model, this called hyperparametering:
### Build Random Forest Model with hyperparameters:
##### Grid Search
In this grid search I will try different combinations of RF hyperparameters.

Most important hyperparameters of Random Forest:

>n_estimators = n of trees

>max_features = max number of features considered for splitting a node

>max_depth = max number of levels in each decision tree

>min_samples_split = min number of data points placed in a node before the node is split

>min_samples_leaf = min number of data points allowed in a leaf node

>bootstrap = method for sampling data points (with or without replacement)


```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [2,4]
# Minimum number of samples required to split a node
min_samples_split = [2, 5]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
# Create the param grid
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(param_grid)
```

Creating the Model

```python
from sklearn.ensemble import RandomForestClassifier
rf_Model = RandomForestClassifier()
```

Creating the Search Grid using our model

```python
from sklearn.model_selection import GridSearchCV
rf_Grid = GridSearchCV(estimator = rf_Model, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4)
```
